# ***Matplotlib***

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
## Plotting X and Y graph


x = [1,2,3]
y = [4,5,6]

plt.plot(x,y)
plt.show()                            ##IT IS NECESSARY TO WRITE THIS LINE SO THAT WE CAN SHOW THE GRAPH IN OTHER PLATFORMS THAN COLAB
```

```python
## Marker has different styles like 'o','D','1', etc and linestyle alse has different types as '-.',-,'--',' ', etc

plt.plot(x,y,marker = 'o',linestyle='--')
plt.show()
```

```python
## USING ' ', OR 'none' TO ERASE LINE 

plt.plot(x,y,marker = 'o',linestyle=' ')
plt.show()
```

```python
plt.plot(x,y,marker = '^',linestyle='dotted',color='LIMEGREEN')
plt.show()
```

```python
## MS= mraker size, MFC= marker face color, MEC= marker edge color

plt.plot(x,y,marker='o',linestyle='dashed',ms=30,mfc='red',mec='black')
plt.show()
```

###WE CAN DO MANY THINGS IN THE GRAPH LIKE MARKERSIZE, LINEWIDTH ETC.


> https://matplotlib.org/2.0.2/api/pyplot_api.html

: : This url contains all the info about the syntax and things that we can modify



```python
## Here ls is equal to linestyl and c = color, just like that linewidth can be written as lw

plt.plot(x,y,marker='o',ls='dashed',c='limegreen',ms=30,mfc='red',mec='black',linewidth=3)
plt.show()
```

```python
## mew= marker edge width

plt.plot(x,y,marker='o',ls='dashed',c='limegreen',ms=30,mfc='red',mec='black',mew=5,linewidth=3)
plt.show()
```

# **SUBPLOTS**



> SubPlots help in plotting more than one graphs together



```python
## plt.subplot has the path or coordinates inside its(1,2,1); It shows that it will have 1 row that would contain 2 columns
## and place the subplot that is below on the first position i.e. plot([1,2,3],[7,8,9]) is in first position

plt.subplot(1,2,1)
plt.plot([1,2,3],[7,8,9])
plt.title('Pepcoding')
plt.subplot(1,2,2)
plt.plot([4,5,6],[7,8,9])
plt.title('Nados')
plt.show()
```

```python
## We have 4 plots that are arranged using sub plot so that we can see that easily


plt.figure(figsize=(19, 6))                                      ## It'll increase the figure size of the graphs


##Scatter plot b/w x and y

plt.subplot(2,3, 1)                                               ## 2 rows, 3 columns and position 1
x = np.round(np.random.rand(7)*7)
y = np.round(np.random.rand(7)*9)

plt.scatter(x, y, s=[1120, 1100, 1500, 1300, 1500, 1800, 2000], alpha=0.6, color = ['r', 'g', 'b', 'y', 'r', 'b', 'b'])


## Plot b/w x^ something

plt.subplot(2,3, 3)                                               ## 2 rows, 3 columns and position 3
x = np.arange(2, 20)
plt.scatter(x, x**2, color = 'r', label='x^2', marker = '^')
plt.scatter(x, x**2.5, color = 'g', label='x^2.5', marker = 'o')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('x * y')
plt.legend()


## Plot between two simple x and y values

plt.subplot(2,3,5)                                                ## 2 rows, 3 columns and position 5
plt.scatter([3, 7, 5], [2, 4, 8])
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Scatter')

plt.subplot(2,3,4)                                                ## 2 rows, 3 columns and position 4
plt.scatter([3, 7, 5], [2, 4, 8])
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Scatter')


plt.show()
```

#**BAR GRAPH**

```python
## plt.bar tp create vertical bar chart and plt.hbar is used for horizontal bar chart

plt.bar(['Jan','Feb','Mar','Apr'],[10,20,6,22],alpha=0.6,color=['r','g','b','y'],width=[0.4,0.6,1,0.9])
            #X-axis values          #Y-axis
            
plt.xlabel('Months')
plt.ylabel('Crime')
```

```python
## here height is used instead of width

plt.barh(['Jan','Feb','Mar','Apr'],[10,20,6,22],alpha=0.6,color=['r','g','b','y'],height=[0.4,0.6,1,0.9])

plt.xlabel('Months')
plt.ylabel('Crime')
plt.show()
```

##**PI CHART**

```python
## The explode part helps in the segments that seems seperated out of the graph that makes it more interactive
## shadow = true helps in creating a blemish shadow under each segment

plt.pie([7, 8, 10, 12, 5, 2], labels = ['N', 'O', 'co2', 'co', 'mg', 'others'], explode = [0.1, 0.2, 0.3, 0.3, 0.2, 0.1], shadow=True)

          #Values that are                # labels of the values that are                                        
          #automatically                  # given respectively and the 
          #converted into %               # color is choosen by the code
plt.show()
```

```python
plt.pie([7, 8, 10, 12, 5, 2], labels = ['N', 'O', 'co2', 'co', 'mg', 'others'], explode = [0, 0, 0, 0.3, 0.2, 0.1], shadow=True)
plt.show()

## Here we can see that by reducing explode to zero the segments of 'N','O','CO2' is reverted back to the original circle
```

```python
## Scatter plot of the setosa df with petal-length and petal_width as x and y coordinates

plt.scatter(setosa['petal_length'],setosa['petal_width'],label = 'setosa', marker="D")
plt.legend()
plt.show()
```

```python
## Plotting all the scatter  plots together.
## The insights are that: if we know the petal length and width of a flower and we plot it on this scatter plot so we can easily guess the
## species that the flower will belong to, just like this we have to do this with machine and that'll be called  ** MACHINE LEARNING **

plt.scatter(setosa['petal_length'],setosa['petal_width'],label = 'setosa', marker="D")
plt.scatter(versicolor['petal_length'],versicolor['petal_width'],label = 'versicolor', marker='o')
plt.scatter(virginica['petal_length'],virginica['petal_width'],label = 'virginica', marker='^')
plt.legend()
plt.show()
```
