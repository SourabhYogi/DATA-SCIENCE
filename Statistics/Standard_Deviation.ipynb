```python
df=pd.read_csv('mtcars.csv')
df.head()
```

```python
sns.boxplot(data=df, x=df['hp'])
```

```python
df['hp'].describe()
```

```python
q1=df['hp'].quantile(0.25)
q3=df['hp'].quantile(0.75)
```

```python
q1
```

```python
q3
```

```python
IQR=q3-q1
IQR
```

## OUTLIERS USING IQR

```python
upper_bound = q3 + (1.5 * IQR)
lower_bound = q1 - (1.5 * IQR)
```

```python
upper_bound
```

```python
lower_bound
```

```python
## Checking the outlier in hp in our data

filter = df['hp'] > upper_bound
df[filter]
```

```python
##Storing every value other than outlier in data (WE CAN SEE THAT 30 IS MISSING )

filt = df['hp'] < upper_bound   
df2=df[filt]
df2
```
