##MISSING DATA

```python
## WE'LL CREATE A DF FROM DICTIONARIES

dict1 = {'A': 48, 'C':69}
dict2 = {'B':5.8, 'C':6.2}
dict3 = {'A': 10, 'B':15, 'C':12}
```

```python
df6 = pd.DataFrame({"Height":dict1,"Weight":dict2,"Age":dict3})
df6                                                                        ##NaN represents Not a Number
```

```python
df6.info
```

```python
## IT'LL DELETE THE COLUMNS THAT HAVE NaN VALUES IN THEM (We've not used inplace) 

df6.dropna(axis=1)
```

```python
df6
```

##How to fill a NaN values 
>.fillna()

```python
## It'll fill the NaN value in height column with the mean of height column

df6['Height'].fillna(df6['Height'].mean(),inplace=True)
df6
```

```python
df6.loc['A']=df6.loc['A'].fillna(30)

df6
```

```python
df6.loc['A'] = df6.loc['A'].fillna(3)
```

```python
df6
```

##HOW TO CONVERT CSV DATA FILE INTO DF IN PANDAS

```python
df10 = pd.read_csv("https://raw.githubusercontent.com/neylsoncrepalde/projeto_eda_covid/master/covid_19_data.csv")
```

```python
df10
```

```python
df10.head()
```

```python
df10.info()
```

```python
df10.shape
```

```python
df10.describe()
```

```python
df10.drop(columns = ['SNo','Last Update'],inplace = True)
```

```python
df10
```

```python
## To rename the columns with something


df10.rename(columns = {"Province/State":"State","Country/Region":"Country"},inplace = True)
```

```python
df10
```

```python
## TO CHANGE THE DATA TYPE OF OBSERVATION DATE TO DATETIME

df10['ObservationDate'] = pd.to_datetime(df10['ObservationDate'])
```

```python
df10
```

```python
## TO GIVE THE MAX VALUE IN THE CONFIRMED COLUMN

max = df10['Confirmed'].max()

df10[df10['Confirmed']==max]
```

## TO GROUP A GIVEN NUMBERS OF ROWS FROM THE DATA

```python
df10.groupby('Country').sum()
```

```python
## TO  GROUP FROM COUNTRY NAME WITH INDEXES

df10.groupby('Country').sum().reset_index()
```

```python
## TO GRROUP FROM COUNTRY AND ONLY SHOWING CONFIRMED AND DEATHS

df10.groupby('Country')[['Confirmed','Deaths']].sum().reset_index()
```

## TO MERGE OR JOIN TWO DF IN PANDAS

```python
## TO MAKE A NEW DF WITH Observation date, state and country

A = df10[ ['ObservationDate', 'State', 'Country'] ]
A.head()
```

```python
B = df10[ ['Confirmed', 'Deaths','Recovered'] ]
B.head()
```

```python
## MERGING A WITH B KEEPING BOTH THEIR INDEXEX AS TRUE OR RIGHT
#1.

A.merge(B, left_index=True, right_index=True)
```

```python
## MERGING A WITH B KEEPING BOTH THEIR INDEXEX AS TRUE OR RIGHT
#2.

A.join(B)
```
