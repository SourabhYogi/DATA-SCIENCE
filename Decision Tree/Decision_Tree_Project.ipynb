```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/shivendutyagi/Datasets/main/mushrooms.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
for column in df.columns:
  print('Unique values in',column+':',
        df[column].unique())
```

```python
X=df.drop('class',axis=1)
y=df['class']
```

```python
## Changing all the string or object values into integer by using "get dummies", that is the function of pandas that will encode the values


X_encoded=pd.get_dummies(X,drop_first=True)
```

```python
X_encoded
```

```python
## Changing the values of y as well, we didn't use get dummies cause it'll creat a seperate column for 'e' and 'p', so we did this

y[y=='p']=1
y[y=='e']=0
```

```python
## Changing the data type of Y from object to int, so we can use decision tree 

y=y.astype('int')

y
```

```python
## Training the model

from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(X_encoded,y,test_size=0.34,random_state=42)
```

```python
X_train
```

```python
## Applying Decision Tree

from sklearn.tree import DecisionTreeClassifier
classifier=DecisionTreeClassifier()
dt=classifier.fit(X_train,y_train)
```

```python
classifier.score(X_train,y_train)
```

```python
classifier.score(X_test,y_test)
```

```python
from sklearn.tree import plot_tree
from sklearn.model_selection import cross_val_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import plot_confusion_matrix
import matplotlib.pyplot as plt
```

```python
## Tree

plt.figure(figsize=(24,10))
plot_tree(dt,filled=True,class_names=['Edible','Poisonous'],feature_names=X_encoded.columns);
```

```python
score=cross_val_score(dt,X_train,y_train,cv=5)
for scores in score:
  print('Accuracy:',round(scores,4))
```

```python
## Confusion Matrix

plot_confusion_matrix(dt,X_test,y_test,display_labels=['Edible','Poisonous'])
plt.show()
```
