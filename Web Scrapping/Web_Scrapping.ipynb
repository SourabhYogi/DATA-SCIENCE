```python
from bs4 import BeautifulSoup                   ##Beautiful Soup is a library
```

```python
import requests                                 ## Requests is used to access and store the web content
```

```python
##requests.get is used to get or store the data on a certain web addess

flipkart = requests.get("https://www.flipkart.com/search?q=apple+mobiles")      
```

```python
## this line is used to convert the web content that was in byte or str form into html parser,

## BeautifulSoup can be used to convert  into xml or other forms

soup = BeautifulSoup(flipkart.text, "html.parser")
```

```python
## To prettify the content 

print(soup.prettify)
```

```python
##To Show the the title of the tab

soup.title
```

```python
soup.title.text
```

```python
## To display the div that are in the page

soup.div.text
```

```python
## Find function gives us the value of the first thing that comes up in classs
## class_ is used cause class is already a key word

mobile = soup.find("div", class_="_13oc-S")
```

```python
mobile.text
```

```python
## find_all is used to store all the things that are in that web page
## This clas  _13oc-S is the div that contains all the displayed mobiles i.e. 24

mobiles = soup.find_all("div", class_="_13oc-S")
```

#**To input the page content into a pandas DataFrame**

```python
## TO CREATE 3 LISTS NAMESLY names, prices & ROMs AND APPLYING LOOP TO FILL THE LISTS WITH VALUES FROM THE PREVIOUSLY USED 
## name, price & ROM 

names = []
prices = []
ROMs = []

for first in mobiles:
    name = first.find("div", attrs = {"class": "_4rR01T"}).text
    price = first.find("div", class_= "_30jeq3 _1_WHN1").text
    ROM = first.find("li", class_= "rgWa7D").text
    

    names.append(name)
    prices.append(price)
    ROMs.append(ROM)
```

```python
type(names)
```

```python
ROMs
```

```python
## Creating a dictionary and storing the lists from above in it so that we can store them in pandas DF 

dict1 = {"Model": names, "Price": prices, "ROM": ROMs}
```

```python
## Creating a DF and storing dict1 in it

import pandas as pd
DF= pd.DataFrame(dict1)
```
