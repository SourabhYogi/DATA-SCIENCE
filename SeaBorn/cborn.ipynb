```python
import seaborn as sns
```

```python
df=sns.load_dataset("tips")
```

```python
df
```

## **REL PLOT**

```python
sns.relplot(x='total_bill',y='tip',data=df)
```

```python
## This hue is a prop that will differntiate the "sex" column with different colors

sns.relplot(x='total_bill',y='tip',data=df,hue='sex')
```

```python
## This style feature will differntiate on the basis of the marks that are used in data point from the column "time"

sns.relplot(x='total_bill',y='tip',data=df,hue='sex',style="time")
```

```python
## This size feature will differntiate according to the size of points from the column "size"

sns.relplot(x='total_bill',y='tip',data=df,hue='sex',style="time",size="size")
```

##Two graphs on the same dataset

```python
## Two graphs that will be based on the categories in the "time" i.e. lunch & dinner

sns.relplot(x="total_bill", y="tip", data=df, hue='smoker',col='time')
```

## **CAT PLOT**

```python
sns.catplot(x='day',y='total_bill',data=df)
```

```python
## In this sns.catplot the jitter is default true; jitter helps us in diiferntiating points on the plot

sns.catplot(x='day',y='total_bill',data=df,jitter=False)
```

```python
sns.catplot(x='smoker',y='tip',data=df)
```

```python
## ordering the data set accprding tp our need

sns.catplot(x='smoker',y='tip',data=df,order=['No','Yes'])
```

### Swarm plot through cat plot

```python
sns.catplot(x='day',y='total_bill',data=df,kind='swarm',hue="size")
```

## **BOX Plot through Catplot**

```python
sns.catplot(x='day',y='tip',kind='box',data=df)
```

```python
sns.catplot(x='day',y='tip',kind='boxen',data=df)
```

```python
sns.catplot(y='day',x='total_bill',data=df,kind='violin',hue="sex", split=True)
```

###Plotting two graphs overlapping one another

```python
## here color='k' makes the color black
## here inner = None will remove the line inside of the violin plot

sns.catplot(x='day',y='total_bill',kind='violin',data=df,inner=None)
sns.swarmplot(x='day',y='total_bill',data=df,color='k',size=3)
```

## **Joint Plot**

```python
x=df['total_bill']
y=df['tip']
```

```python
sns.jointplot(x=x,y=y)
```

```python
## Another way for Joint Plot

sns.jointplot(x='total_bill',y='tip', data=df)
```

```python
## HEX plot in Joint Plot

sns.jointplot(x=x,y=y,kind='hex')
```

```python

sns.jointplot(x=x,y=y,kind='hex',cmap='Reds')
```

```python
## KDE plot through joiplot

sns.jointplot(x=x,y=y,kind='kde', cmap='BrBG')
```

## **REGRESSION PLOT**

* Reg Plot
* LM Plot

```python
## BOTH THE PLOTS ARE ALMOST SIMILAR

sns.regplot(x='total_bill',y='tip',data=df, scatter_kws={'color':'Green'}, line_kws={'color':"Crimson"})
plt.show()

## THE LINE IS THE BEST FIT LINE
```

```python
sns.lmplot(x='total_bill',y='tip',data=df, scatter_kws={'color':'cyan'}, line_kws={'color':"magenta"})
plt.show()
```
