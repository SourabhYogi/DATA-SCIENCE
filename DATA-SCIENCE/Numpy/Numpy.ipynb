### To show that list creates a copy when we access elements and array simply shows the elements and doesn't create a copy thus saving time

```python
lst = [1,2,3,4,5]
lst2 = lst[2:4]
print(lst)
print(lst2)
lst2[0] = 1000
print(lst)
print(lst2)
```

```python
## It shows that the value of original arr12 also changes, as it only shows us the original array's value
arr12 = np.array([1,2,3,4,5])
arr13 = arr12[2:4]
print(arr12)                              
print(arr13)
arr13[0] = 1000
print(arr12)
print(arr13)
```

```python
arr14 = np.array([1,2,3,4,5])
arr15 = arr14[2:4].copy()           ##arr.copy to create a copy array arr15 with 2nd & 4th element of arr14
print(arr14)
print(arr15)
arr15[0] = 1000
print(arr14)
print(arr15)
```

#Operations On Array

```python
a = np.array([1,2,34,4,5])
b = np.array([6,7,8,9,10])
```

```python
## (a+b) will add the element of a with the same element of b (ex. a[0,0] with b[0,0] )

print(a+b)
print(a-b)
print(a*b)
print(a/b)
```

```python
## It will return true where the condition is right and false where it is false

print(a>b)
print(a<b)
print(a==b)
print(a>=b)
print(a<=b)
```

```python
## SOME MORE FUNCTIONS OF ARRAY

a29=np.array([1,2,3,4,5,6,7,8])
a29
```

```python
## SOME MORE FUNCTIONS OF ARRAY

print(a29.mean())              ## Mean of array  
print(a29.sum())               ## Sum of array
print(a29.std())               ## Standard Deviation of the array

print(a29.max())               ## This will give us the max value in the array 
print(a29.min())               ## This will give us the min value in the array

print(a29.argmin())            ## This will give us the index of the min value in the array
print(a29.argmax())            ## This will give us the index of the max value in the array


```

# **CONCATENATE**

---

### * Very important function, it is used to merge two arrays into one




###**SPLIT**
### *  Its a function to split a array into smaller arrays





```python
a = np.array([
              [1,2,3],
              [4,5,6]
])
b = np.array([
              [7,8,9]
])


ar5= np.concatenate((a,b), axis=0)      ## axis=o shows that the merging will be along the rows
print(ar5)
```
